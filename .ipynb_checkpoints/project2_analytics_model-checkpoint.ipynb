```python
import pandas as pd
import numpy as np
from pathlib import Path

from IPython.display import display

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-darkgrid")
RANDOM_STATE = 42
DATA_PATH = Path("/Users/choseoyeon/Desktop/umich/670/project/clean_posts_ready.csv")
```

```python
df = pd.read_csv(DATA_PATH)

initial_rows = len(df)
missing_target = df["engagement_rate"].isna().sum()

print(f"원본 샘플 수: {initial_rows:,}, 피쳐 수: {df.shape[1]}")
if missing_target:
    print(
        f"engagement_rate 결측 행 {missing_target:,}개 제거 -> {initial_rows - missing_target:,}개 활용"
    )

df = df.dropna(subset=["engagement_rate"]).copy()

print(f"모델링용 샘플 수: {len(df):,}, 피쳐 수: {df.shape[1]}")
display(df.head(3))

engagement_summary = df["engagement_rate"].describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95])
missing_snapshot = (
    df.isna()
      .mean()
      .sort_values(ascending=False)
      .head(10)
      .to_frame(name="missing_ratio")
)

print("\nengagement_rate 분포 요약:")
display(engagement_summary)

print("\n결측치 비율 상위 10개 피쳐:")
display(missing_snapshot)
```

```python
EXCLUDED_FEATURES = [
    "engagement_rate",
    "engagement_score",
    "like_count",
    "comment_count",
    "like_to_comment_ratio"
]

numeric_df = df.select_dtypes(include=[np.number]).copy()
X = numeric_df.drop(columns=[col for col in EXCLUDED_FEATURES if col in numeric_df.columns])
y = df["engagement_rate"].astype(float)

print(f"사용 가능한 수치 피쳐 수: {X.shape[1]}")
feature_preview = pd.Series(X.columns).sort_values().head(15)
print("피쳐 예시:")
display(feature_preview)

ranked = y.rank(method="first")
stratify_labels = pd.qcut(ranked, q=5, labels=False)

X_train, X_test, y_train_reg, y_test_reg = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=stratify_labels
)

median_threshold = y_train_reg.median()
y_train_clf = (y_train_reg >= median_threshold).astype(int)
y_test_clf = (y_test_reg >= median_threshold).astype(int)

class_balance = (
    pd.DataFrame({
        "count": y_train_clf.value_counts().sort_index(),
        "ratio": y_train_clf.value_counts(normalize=True).sort_index()
    })
    .rename(index={0: "낮음", 1: "높음"})
)

print(f"훈련 세트 중앙값 기준: {median_threshold:.4f}")
print("로지스틱 회귀용 클래스 분포:")
display(class_balance)
```

```python
logistic_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=2000,
        class_weight="balanced",
        random_state=RANDOM_STATE
    ))
])

logistic_pipeline.fit(X_train, y_train_clf)

y_pred_clf = logistic_pipeline.predict(X_test)
y_prob_clf = logistic_pipeline.predict_proba(X_test)[:, 1]

report_dict = classification_report(
    y_test_clf,
    y_pred_clf,
    target_names=["낮음", "높음"],
    output_dict=True
)
report_df = pd.DataFrame(report_dict).T
roc_auc = roc_auc_score(y_test_clf, y_prob_clf)

print("로지스틱 회귀 베이스라인 성능 (테스트 세트)")
display(report_df)
print(f"ROC-AUC: {roc_auc:.3f}")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
ConfusionMatrixDisplay.from_predictions(
    y_test_clf,
    y_pred_clf,
    display_labels=["낮음", "높음"],
    cmap="Blues",
    ax=axes[0],
    colorbar=False
)
axes[0].set_title("로지스틱 회귀 혼동 행렬")

RocCurveDisplay.from_predictions(
    y_test_clf,
    y_prob_clf,
    ax=axes[1],
    name="Logistic"
)
axes[1].set_title("로지스틱 ROC 곡선")
plt.tight_layout()
plt.show()

coef_series = pd.Series(
    logistic_pipeline.named_steps["model"].coef_[0],
    index=X_train.columns
).sort_values(key=lambda s: s.abs(), ascending=False)

print("영향력이 큰 상위 10개 피쳐 (절댓값 기준)")
display(coef_series.head(10).to_frame(name="coefficient"))
```

```python
tree_clf_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", DecisionTreeClassifier(
        max_depth=8,
        min_samples_leaf=60,
        class_weight="balanced",
        random_state=RANDOM_STATE
    ))
])

tree_clf_pipeline.fit(X_train, y_train_clf)

y_pred_tree_clf = tree_clf_pipeline.predict(X_test)
y_prob_tree_clf = tree_clf_pipeline.predict_proba(X_test)[:, 1]

tree_clf_report = classification_report(
    y_test_clf,
    y_pred_tree_clf,
    target_names=["낮음", "높음"],
    output_dict=True
)
tree_clf_report_df = pd.DataFrame(tree_clf_report).T
tree_clf_roc_auc = roc_auc_score(y_test_clf, y_prob_tree_clf)

print("의사결정나무 분류 성능 (테스트 세트)")
display(tree_clf_report_df)
print(f"ROC-AUC: {tree_clf_roc_auc:.3f}")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
ConfusionMatrixDisplay.from_predictions(
    y_test_clf,
    y_pred_tree_clf,
    display_labels=["낮음", "높음"],
    cmap="Purples",
    ax=axes[0],
    colorbar=False
)
axes[0].set_title("의사결정나무 혼동 행렬")

RocCurveDisplay.from_predictions(
    y_test_clf,
    y_prob_tree_clf,
    ax=axes[1],
    name="Decision Tree"
)
axes[1].set_title("의사결정나무 ROC 곡선")
plt.tight_layout()
plt.show()

importance_clf = pd.Series(
    tree_clf_pipeline.named_steps["model"].feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

print("의사결정나무 분류 중요 피쳐 상위 10개")
display(importance_clf.head(10).to_frame(name="importance"))
```

```python
tree_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("model", DecisionTreeRegressor(
        max_depth=10,
        min_samples_leaf=50,
        random_state=RANDOM_STATE
    ))
])

upper_quartile = y_train_reg.quantile(0.75)
sample_weight_train = np.where(y_train_reg.values >= upper_quartile, 3.0, 1.0)

tree_pipeline.fit(X_train, y_train_reg, model__sample_weight=sample_weight_train)

y_pred_reg = tree_pipeline.predict(X_test)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred_reg)

print("의사결정나무 회귀 성능 (테스트 세트)")
print(f"MAE : {mae:.5f}")
print(f"RMSE: {rmse:.5f}")
print(f"R^2 : {r2:.3f}")

importances = pd.Series(
    tree_pipeline.named_steps["model"].feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

top_features = importances.head(10)
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=top_features, y=top_features.index, palette="viridis", ax=ax)
ax.set_title("의사결정나무 중요 피쳐 상위 10개")
ax.set_xlabel("importance")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(y_test_reg, y_pred_reg, alpha=0.4)
ax.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], "r--", lw=2)
ax.set_xlabel("실제 engagement_rate")
ax.set_ylabel("예측 engagement_rate")
ax.set_title("의사결정나무 예측 vs 실제")
plt.tight_layout()
plt.show()
```

### 노트
- engagement_rate는 훈련 세트 중앙값 이상/미만으로 이진 분류하여 로지스틱 회귀와 의사결정나무 분류 모델을 구성했고, 모두 `class_weight='balanced'` 혹은 가중치 전략을 넣어 불균형을 완화했습니다.
- 좋아요/댓글 및 해당 파생 피쳐(`like_count`, `comment_count`, `engagement_score`, `like_to_comment_ratio`, `engagement_rate`)는 모두 입력 피쳐에서 제외했습니다.
- 로지스틱/의사결정나무 분류 셀에서 `classification_report` 표, 혼동 행렬, ROC 곡선을 한 번에 확인할 수 있으며, 기준선 대비 결정트리의 개선 여부를 시각적으로 비교할 수 있습니다.
- 의사결정나무 회귀에는 상위 25% 구간 샘플에 더 큰 가중치(3.0)를 부여해 희소한 고 engagement 사례를 강조했습니다.
- 모든 파이프라인이 `SimpleImputer`로 결측치를 처리하며, 로지스틱 모델에는 `StandardScaler`를 포함해 안정적으로 학습하도록 했습니다. 필요 시 나무 깊이, 리프 크기, 샘플 가중치 등을 조정해 추가 실험을 진행할 수 있습니다.
