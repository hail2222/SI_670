# Project 2 Analytics

clean_posts_ready.csv 기반으로 광고 여부, 군집 구조, 참여도 관련 특징을 심층 분석한다.


## 분석 목표
- 전체 포스트 클러스터링 및 클러스터 특징 요약
- 광고(ad) vs 비광고(non-ad) 비교 및 참여도 차이 검정
- 전체/광고 서브셋에서 engagement rate와 상관성이 높은 피처 탐색
- 광고/비광고 텍스트 특징 차이와 고참여 텍스트 특성 파악


```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False


pd.set_option("display.max_columns", None)
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
DATA_PATH = Path("clean_posts_ready.csv")

if not DATA_PATH.exists():
    raise FileNotFoundError(f"데이터 파일이 존재하지 않습니다: {DATA_PATH.resolve()}")

raw_df = pd.read_csv(DATA_PATH)
raw_df["timestamp"] = pd.to_datetime(raw_df["timestamp"], errors="coerce")

binary_cols = [
    "is_video",
    "owner_is_verified",
    "has_hashtag",
    "contains_mention",
    "has_url",
    "is_ad_text",
]
for col in binary_cols:
    raw_df[col] = raw_df[col].fillna(0).astype(int)

raw_df["num_hashtags"] = raw_df["num_hashtags"].fillna(0)
raw_df["num_emojis"] = raw_df["num_emojis"].fillna(0)
raw_df["followers"] = raw_df["followers"].replace({0: np.nan})
raw_df["engagement_rate"] = raw_df["engagement_rate"].astype(float)

raw_df.head()
```

```python
summary = {
    "총 포스트 수": len(raw_df),
    "광고 비중": raw_df["is_ad_text"].mean(),
    "평균 참여율": raw_df["engagement_rate"].mean(),
    "중앙값 참여율": raw_df["engagement_rate"].median(),
}
pd.Series(summary)
```

```python
missing_pct = raw_df.isna().mean().sort_values(ascending=False)
missing_pct.head(10)
```

```python
df = raw_df.dropna(subset=["followers"]).copy()
print(f"followers 정보 누락으로 제외된 포스트 수: {len(raw_df) - len(df)}")

text_cols = ["caption_text_clean", "comments_clean"]
for col in text_cols:
    df[col] = df[col].fillna("")

df["log_followers"] = np.log1p(df["followers"].fillna(0))
df["log_like_count"] = np.log1p(df["like_count"].fillna(0))
df["log_comment_count"] = np.log1p(df["comment_count"].fillna(0))
df["like_comment_gap"] = df["like_count"].fillna(0) - df["comment_count"].fillna(0)
df["has_caption"] = (df["caption_length"].fillna(0) > 0).astype(int)
df["has_comments"] = (df["comment_count"].fillna(0) > 0).astype(int)
df["hashtags_per_100_words"] = np.where(
    df["caption_length"] > 0,
    df["num_hashtags"] / (df["caption_length"] / 100),
    0,
)
df["emoji_per_100_words"] = np.where(
    df["caption_length"] > 0,
    df["num_emojis"] / (df["caption_length"] / 100),
    0,
)
sentiment_map = {"negative": -1, "neutral": 0, "positive": 1}
df["caption_sentiment_num"] = df["caption_sentiment"].map(sentiment_map)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
df.head()
```

```python
cluster_features = [
    "log_like_count",
    "log_comment_count",
    "log_followers",
    "engagement_rate",
    "num_hashtags",
    "num_emojis",
    "caption_length",
    "has_hashtag",
    "contains_mention",
    "has_url",
    "is_video",
    "is_ad_text",
    "owner_is_verified",
    "hashtags_per_100_words",
    "emoji_per_100_words",
]

scaler = StandardScaler()
X = scaler.fit_transform(df[cluster_features])

silhouette_results = []
for k in range(2, 9):
    model = KMeans(n_clusters=k, random_state=42, n_init=25)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_results.append({"k": k, "silhouette": score})

silhouette_df = pd.DataFrame(silhouette_results)
silhouette_df
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=silhouette_df, x="k", y="silhouette", marker="o")
plt.title("Silhouette Score by K")
plt.show()
```

```python
best_k = int(silhouette_df.loc[silhouette_df["silhouette"].idxmax(), "k"])
cluster_model = KMeans(n_clusters=best_k, random_state=42, n_init=25)
df["cluster"] = cluster_model.fit_predict(X)

print(f"선택된 클러스터 수: {best_k}")
```

```python
cluster_summary = (
    df.groupby("cluster").agg(
        표본수=("cluster", "size"),
        광고비중=("is_ad_text", "mean"),
        평균참여율=("engagement_rate", "mean"),
        중간참여율=("engagement_rate", "median"),
        평균팔로워=("followers", "mean"),
        평균좋아요=("like_count", "mean"),
        평균댓글=("comment_count", "mean"),
        평균해시태그수=("num_hashtags", "mean"),
        평균이모지수=("num_emojis", "mean"),
    )
    .sort_values("평균참여율", ascending=False)
)
cluster_summary
```

```python
cluster_feature_means = df.groupby("cluster")[cluster_features].mean()
overall_means = df[cluster_features].mean()
feature_lift = cluster_feature_means - overall_means

feature_lift
```

```python
def describe_cluster(cluster_id: int, top_n: int = 6) -> pd.DataFrame:
    lift = feature_lift.loc[cluster_id]
    top_features = lift.abs().sort_values(ascending=False).head(top_n).index
    return pd.DataFrame({"lift": lift[top_features]}).sort_values("lift", ascending=False)

cluster_descriptions = {cid: describe_cluster(cid) for cid in sorted(df["cluster"].unique())}
cluster_descriptions
```

```python
cluster_plot_df = (
    df.groupby("cluster").agg(
        avg_engagement_rate=("engagement_rate", "mean"),
        post_count=("cluster", "size"),
    )
    .reset_index()
)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.barplot(data=cluster_plot_df, x="cluster", y="post_count", ax=ax[0])
ax[0].set_title("Post counts per cluster")
sns.barplot(data=cluster_plot_df, x="cluster", y="avg_engagement_rate", ax=ax[1])
ax[1].set_title("Avg engagement rate per cluster")
plt.tight_layout()
plt.show()
```

```python
ad_group_summary = (
    df.groupby("is_ad_text").agg(
        표본수=("is_ad_text", "size"),
        평균참여율=("engagement_rate", "mean"),
        중앙참여율=("engagement_rate", "median"),
        평균좋아요=("like_count", "mean"),
        평균댓글=("comment_count", "mean"),
        평균팔로워=("followers", "mean"),
        평균해시태그수=("num_hashtags", "mean"),
        평균이모지수=("num_emojis", "mean"),
        멘션비중=("contains_mention", "mean"),
        링크비중=("has_url", "mean"),
    )
    .rename(index={0: "non-ad", 1: "ad"})
)
ad_group_summary
```

```python
ad_eng = df.loc[df["is_ad_text"] == 1, "engagement_rate"]
non_ad_eng = df.loc[df["is_ad_text"] == 0, "engagement_rate"]

test_result = stats.ttest_ind(ad_eng, non_ad_eng, equal_var=False)

def cohens_d(x, y):
    nx, ny = len(x), len(y)
    pooled_std = np.sqrt(
        ((nx - 1) * np.var(x, ddof=1) + (ny - 1) * np.var(y, ddof=1)) / (nx + ny - 2)
    )
    return (np.mean(x) - np.mean(y)) / pooled_std

d_result = cohens_d(ad_eng, non_ad_eng)

pd.Series(
    {
        "t-stat": test_result.statistic,
        "p-value": test_result.pvalue,
        "Cohen's d": d_result,
        "ad 평균 참여율": ad_eng.mean(),
        "non-ad 평균 참여율": non_ad_eng.mean(),
    }
)
```

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(data=df, x="engagement_rate", hue="is_ad_text", common_norm=False)
plt.title("Engagement rate distribution by Ad status")
plt.xlabel("engagement_rate")
plt.show()
```

```python
all_corr_candidates = [
    "num_hashtags",
    "num_emojis",
    "caption_length",
    "contains_mention",
    "has_url",
    "has_hashtag",
    "is_video",
    "owner_is_verified",
    "caption_sentiment_num",
    "comments_sentiment_avg",
    "comment_positive_ratio",
    "comment_neutral_ratio",
    "comment_negative_ratio",
    "hashtags_per_100_words",
    "emoji_per_100_words",
    "has_caption",
    "has_comments",
    "is_ad_text",
    "followers",
]

derived_from_formula = {
    "like_count",
    "comment_count",
    "engagement_score",
    "like_to_comment_ratio",
    "log_followers",
    "log_like_count",
    "log_comment_count",
    "like_comment_gap",
    "has_comments"
}

corr_features = [feat for feat in all_corr_candidates if feat not in derived_from_formula]

corr_series = (
    df[corr_features].corrwith(df["engagement_rate"]).sort_values(key=lambda s: s.abs(), ascending=False)
)
corr_series
```

```python
top_corr = corr_series.head(10)
plt.figure(figsize=(6, 4))
sns.barplot(x=top_corr.values, y=top_corr.index)
plt.title("engagement_rate와 가장 상관 높은 피처")
plt.xlabel("상관계수")
plt.show()
```

```python
ad_corr = (
    df.loc[df["is_ad_text"] == 1, corr_features]
    .corrwith(df.loc[df["is_ad_text"] == 1, "engagement_rate"])
    .dropna()
    .sort_values(key=lambda s: s.abs(), ascending=False)
)

ad_corr.head(10)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=ad_corr.head(10).values, y=ad_corr.head(10).index)
plt.title("Engagement_rate correlation in the ad post Top 10")
plt.xlabel("Correlation coefficient")
plt.show()
```

```python
ad_related_terms = {
    "ad",
    "ads",
    "advert",
    "advertise",
    "advertised",
    "advertisement",
    "advertisements",
    "advertising",
    "sponsored",
    "sponsor",
    "sponsorship",
    "promo",
    "promotion",
    "promocode",
    "discount",
    "partnership",
    "branddeal",
    "brand",
}
custom_stop_words = ENGLISH_STOP_WORDS.union(ad_related_terms)

vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    min_df=10,
    stop_words=list(custom_stop_words),
)
tfidf_matrix = vectorizer.fit_transform(df["caption_text_clean"])
terms = vectorizer.get_feature_names_out()

ad_mask = (df["is_ad_text"] == 1).values
ad_mean = tfidf_matrix[ad_mask].mean(axis=0).A1
non_ad_mean = tfidf_matrix[~ad_mask].mean(axis=0).A1

term_scores = pd.DataFrame(
    {
        "term": terms,
        "ad_mean": ad_mean,
        "non_ad_mean": non_ad_mean,
    }
)
term_scores["ad_lift"] = term_scores["ad_mean"] - term_scores["non_ad_mean"]

ad_top_terms = term_scores.sort_values("ad_lift", ascending=False).head(15)
non_ad_top_terms = term_scores.sort_values("ad_lift").head(15)

ad_top_terms
```

```python
non_ad_top_terms
```

```python
high_threshold = df["engagement_rate"].quantile(0.75)
df["high_engagement"] = (df["engagement_rate"] >= high_threshold).astype(int)

{
    "상위 25% 기준": high_threshold,
    "고참여 포스트 수": int(df["high_engagement"].sum()),
}
```

```python
y_high = df["high_engagement"]
log_reg = LogisticRegression(max_iter=2000, class_weight="balanced")
log_reg.fit(tfidf_matrix, y_high)

coef_df = pd.DataFrame({"term": terms, "coef": log_reg.coef_[0]})

high_eng_terms = coef_df.sort_values("coef", ascending=False).head(15)
low_eng_terms = coef_df.sort_values("coef").head(15)

high_eng_terms
```

## 종합 분석 리포트 (followers 포함 최신 실행 기준)

### 데이터 정제 및 분석 단위
원본 2,000개 포스트 중 팔로워 정보가 명확한 1,969개만 사용. 참여율 계산의 분모가 확실한 샘플만 남기며, 이후 모든 통계·모델링이 동일한 집합에서 이뤄집니다.

### 1. 전체 포스트 클러스터링
k=5가 실루엣 0.226으로 최적. 각 클러스터 핵심 특징:
- **C3(195개, 광고 0.5%)**: 팔로워 154만, 좋아요 3만, 참여율 1,300으로 압도적 상위 계층. 짧은 캡션·적은 해시태그로도 반응이 터지는 대형 계정.
- **C0(112개, 광고 100%)**: 팔로워 5.2만, 참여율 537. 긴 캡션(평균 416단어)과 11개 내외 해시태그를 적극 활용하는 광고 셀럽군.
- **C1(528개, 비광고)**: 팔로워 6.8만, 참여율 301. 짧은 캡션(146단어)·거의 없는 해시태그로 핵심 정보만 전달하는 계정들.
- **C2(1,041개, 비광고)**: 팔로워 5.1만, 참여율 175. 긴 캡션과 다해시태그 조합에도 반응이 낮아 “양보다 질”이 필요함을 시사.
- **C4(93개, 광고 14%)**: 팔로워 18만, 참여율 138. 장문 캡션·해시태그 과다 사용으로도 효과가 제한적인 중간 계층.

요약: “초고가치 대형 계정(C3)”과 “장문·해시태그 전략이지만 반응이 낮은 중간 집단(C2/C4)”이 뚜렷이 구분되며, 광고 전용 클러스터(C0)가 별도로 존재.

### 2. 광고 vs 비광고 참여율 비교
- 평균 참여율: 광고 0.492, 비광고 0.328. 표면적 차이는 있으나 Welch t-test p≈0.516, Cohen’s d≈0.043 → 통계적으로 유의하지 않음.
- 광고군은 멘션 비중 74%, URL 포함 10%, 해시태그 11개, 캡션 길이 417단어로 비광고 대비 훨씬 공들인 포맷을 사용하지만, 이를 통한 참여율 상승이 확실치 않다는 점을 확인.

### 3. 전체 상관분석 (followers 포함)
- `owner_is_verified`(+0.066)가 가장 높은 상관: 인증 계정일수록 참여율이 높음.
- `caption_length`(-0.027), `num_hashtags`(-0.019), `num_emojis`(-0.015) 등 “과도한 텍스트·해시태그·이모지”는 오히려 참여율을 깎는 경향.
- 댓글 감정비(`comment_positive_ratio`·`comment_neutral_ratio`)나 `hashtags_per_100_words` 같은 비율 피처도 소폭 영향.
- `followers` 자체는 전체 상관 Top10에는 들지 않을 정도로 영향이 미미하여, 전체 스케일에서는 팔로워 수보다 계정 인증 여부나 콘텐츠 스타일이 더 큰 설명력을 가진다는 결론.

### 4. 광고 서브셋 상관분석
- `caption_sentiment_num`(-0.26), `num_hashtags`(-0.12), `caption_length`(-0.11): 광고 문장이 지나치게 긍정적이거나 장문·다해시태그일수록 참여율이 낮음.
- `emoji_per_100_words`(+0.083), `contains_mention`(+0.083), `comment_positive_ratio`(+0.070): 이모지 밀도, 멘션, 긍정 댓글 비중이 상대적으로 효과적.
- `followers`가 -0.081로 Top10에 다시 등장 → 광고 계정은 팔로워가 많아질수록 참여율이 희석(즉, 분모 효과)되는 현상이 뚜렷함. 대형 광고 계정은 CTA/감성 강화를 통해 분모 부담을 상쇄할 필요가 있음.

### 5. 광고 vs 비광고 텍스트 특징 (광고 관련 단어 stopword 처리)
- 광고군 상위 TF-IDF 단어: `new`, `adventure`, `love`, `check`, `feels`, `travel`, `use code`, `local`, `ve`, `fav`. 광고성 단어를 제거했음에도 여전히 체험 공유, 여행·라이프스타일, CTA(코드 입력) 등이 핵심 키워드.
- 비광고군: `ready`, `happy`, `night`, `birthday`, `thank`, `sunday`, `christmas`, `photo`, `beach`, `tomorrow`. 일상 감성, 주말·기념일, 사진 공유 중심의 자연스러운 스토리텔링이 강조됨.

결론: 광고 텍스트는 CTA·체험 강조 어휘를 강화하되, 비광고 톤에서 보이는 “일상/감성/감사” 요소를 적절히 차용하면 자연스러움을 높일 여지가 있음.

### 6. 고참여 텍스트 신호 (상위 25% 기준 로지스틱)
- 양의 계수: `hello`, `little`, `enjoy`, `thanks`, `miss`, `thankful`, `doing`, `walk`, `great`. 인사, 감정 공유, 활동 묘사 등 관계 중심 단어들이 고참여를 견인.
- 음의 계수: `city`, `cream`, `picoftheday`, `tell`, `awesome`, `thank`, `friday`, `thought`, `finally`, `gorgeous`. 특정 제품/도시·루틴 보고 스타일은 반응이 낮은 편.
- 특히 `thank` 단어가 고참여(+1.31)와 저참여(-1.20)에 동시에 등장하는데, 문맥에 따라 “감사 인사(긍정)” vs “루틴적인 해시태그(#thankful, #thankgod 등)의 남용”을 구분할 필요가 있습니다.

### 종합 인사이트
- 대형 계정(C3)과 소/중형 광고 계정(C0/C4) 간 전략 차별화가 필요. 광고 계정은 텍스트 길이/해시태그 양만 늘리기보다 멘션·이모지 등 CTA 장치를 섬세하게 조정해야 함.
- 참여율은 팔로워·좋아요·댓글 수보다는 “콘텐츠 톤과 계정 신뢰 지표(인증)”에 더 민감. 특히 광고에서는 긍정 일변도의 문장보다 진솔한 톤이 유리할 수 있음.
- 텍스트 측면에서는 CTA 단어와 일상 감성 단어 사이의 균형이 중요. 고참여 키워드(`hello`, `enjoy`, `thanks` 등)를 참고해 광고 문구를 인간적인 톤으로 재구성하면 성과 개선 여지가 큼.



## 광고 포스트 효율 인사이트
- **규모보다 톤**: 광고 서브셋 상관에서 `followers`가 음(-0.08)으로 나타났듯, 팔로워 규모가 클수록 참여율이 자동으로 높아지지 않는다. 대형 계정일수록 분모 부담을 감당할 메시지 설계가 필요하다.
- **감성 밸런스 조정**: `caption_sentiment_num`이 -0.26으로 강한 음의 상관을 보여 지나치게 긍정 일변도인 문장은 반응을 떨어뜨릴 수 있다. 제품 찬양보다 공감형 서사를 섞어 진솔한 톤을 유지하는 편이 유리하다.
- **장문·다해시태그 남용 자제**: `caption_length`(-0.11)와 `num_hashtags`(-0.12)가 광고 포스트에서 모두 음의 효과를 냈다. 필수 정보와 CTA에 집중해 짧고 선명한 카피를 쓰는 것이 성과 측면에서 우월하다.
- **CTA 장치(이모지·멘션) 적극 활용**: `emoji_per_100_words`(+0.08), `contains_mention`(+0.08)이 양의 상관을 보여, 광고임을 숨기기보다 이모지·멘션으로 구체적 행동을 자연스럽게 유도하는 접근이 효과적이다.
- **텍스트 키워드 전략**: 광고 상위 단어(`new`, `adventure`, `use code`, `local`)와 고참여 키워드(`hello`, `enjoy`, `thanks`, `miss`, `thankful`) 조합은 “체험+관계 구축” 메시지가 참여율을 올린다는 뜻이다. CTA 단어만 나열하지 말고 인간적인 인사/감사 표현을 섞어라.
- **댓글 분위기 관리**: `comment_positive_ratio`(+), `comment_neutral_ratio`(-)가 상위에 올라왔으므로, 댓글 피드백을 긍정적으로 유지·확산시키는 커뮤니티 관리가 광고 참여율 개선에 기여한다.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

high_cutoff = df["engagement_rate"].quantile(0.75)
df["high_engagement"] = (df["engagement_rate"] >= high_cutoff).astype(int)
print({"high_engagement 비율": df["high_engagement"].mean(), "cutoff": high_cutoff})

exclude_cols = {
    "engagement_rate",
    "engagement_rate_log",
    "engagement_score",
    "engagement_residual",
    "followers",
    "like_count",
    "comment_count",
    "log_followers",
    "log_like_count",
    "log_comment_count",
    "like_comment_gap",
    "like_to_comment_ratio",
    "has_comments",
}

numeric_cols = df.select_dtypes(include=[np.number]).columns
target_col = "high_engagement"
feature_cols = [col for col in numeric_cols if col not in exclude_cols and col != target_col]

X = df[feature_cols].copy()
y = df[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

log_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=2000, class_weight="balanced"))
])
log_pipeline.fit(X_train, y_train)
y_pred_log = log_pipeline.predict(X_test)
y_prob_log = log_pipeline.predict_proba(X_test)[:, 1]

print("[Logistic Regression] classification report:\n", classification_report(y_test, y_pred_log, digits=3))
print("[Logistic Regression] ROC-AUC:", roc_auc_score(y_test, y_prob_log))

```

```python
tree_params = {
    "max_depth": [4, 6, 8, 10],
    "min_samples_split": [20, 40, 80],
    "min_samples_leaf": [10, 20, 40],
    "criterion": ["gini", "entropy"],
}

tree_grid = GridSearchCV(
    DecisionTreeClassifier(class_weight="balanced", random_state=42),
    param_grid=tree_params,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
)
tree_grid.fit(X_train, y_train)
print("Best DT params:", tree_grid.best_params_)

tree_best = tree_grid.best_estimator_
y_pred_tree = tree_best.predict(X_test)
y_prob_tree = tree_best.predict_proba(X_test)[:, 1]

print("[Decision Tree - tuned] classification report:\n", classification_report(y_test, y_pred_tree, digits=3))
print("[Decision Tree - tuned] ROC-AUC:", roc_auc_score(y_test, y_prob_tree))

feature_importance = pd.Series(tree_best.feature_importances_, index=feature_cols).sort_values(ascending=False)
print("상위 중요 피처:")
print(feature_importance.head(10))

```

### Step 1. Numeric + 텍스트 결합 피처 구성


```python
#%pip install imbalanced-learn
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import precision_recall_curve
from imblearn.over_sampling import SMOTE

text_col = "caption_text_clean"
numeric_transformer = Pipeline([
    ("scaler", StandardScaler())
])
text_transformer = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=2000, stop_words="english")),
    ("svd", TruncatedSVD(n_components=50, random_state=42))
])

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, feature_cols),
    ("text", text_transformer, text_col),
])

X_all = df[feature_cols + [text_col]]
y_all = df[target_col]

X_train_adv, X_test_adv, y_train_adv, y_test_adv = train_test_split(
    X_all, y_all, test_size=0.25, stratify=y_all, random_state=42
)

```

### Step 2. Logistic Regression (with text features)


```python
log_text_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("clf", LogisticRegression(max_iter=3000, class_weight="balanced", solver="saga"))
])
log_text_pipeline.fit(X_train_adv, y_train_adv)
y_prob_log_adv = log_text_pipeline.predict_proba(X_test_adv)[:, 1]
y_pred_log_adv = log_text_pipeline.predict(X_test_adv)

print("[Logistic + text] report:\n", classification_report(y_test_adv, y_pred_log_adv, digits=3))
print("[Logistic + text] ROC-AUC:", roc_auc_score(y_test_adv, y_prob_log_adv))

precision, recall, thresholds = precision_recall_curve(y_test_adv, y_prob_log_adv)
f1_scores = 2 * precision * recall / (precision + recall + 1e-9)
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
print({"best_threshold": best_threshold, "best_F1": f1_scores[best_idx]})

y_pred_log_tuned = (y_prob_log_adv >= best_threshold).astype(int)
print("[Logistic + text @best threshold] report:\n", classification_report(y_test_adv, y_pred_log_tuned, digits=3))

```

### Step 3. HistGradientBoostingClassifier


```python
hgb_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("clf", HistGradientBoostingClassifier(class_weight="balanced", max_depth=10, random_state=42))
])
hgb_pipeline.fit(X_train_adv, y_train_adv)
y_prob_hgb = hgb_pipeline.predict_proba(X_test_adv)[:, 1]
y_pred_hgb = hgb_pipeline.predict(X_test_adv)

print("[HistGradientBoosting] report:\n", classification_report(y_test_adv, y_pred_hgb, digits=3))
print("[HistGradientBoosting] ROC-AUC:", roc_auc_score(y_test_adv, y_prob_hgb))

```

### Step 5. Combined Text + GradientBoosting (GridSearch)


```python
from sklearn.ensemble import GradientBoostingClassifier


combo_preprocessor = ColumnTransformer([
    ("num", numeric_transformer, feature_cols),
    ("text", Pipeline([
        ("tfidf", TfidfVectorizer(max_features=3000, stop_words="english")),
        ("svd", TruncatedSVD(n_components=80, random_state=42))
    ]), "combined_text"),
])

X_combo = df[feature_cols + ["combined_text"]]
y_combo = y_all

X_train_combo, X_test_combo, y_train_combo, y_test_combo = train_test_split(
    X_combo, y_combo, test_size=0.25, stratify=y_combo, random_state=42
)

gb_params = {
    "clf__learning_rate": [0.05, 0.1],
    "clf__max_depth": [3, 5],
    "clf__min_samples_leaf": [20, 40],
}

gb_pipeline = Pipeline([
    ("preprocess", combo_preprocessor),
    ("clf", GradientBoostingClassifier(subsample=0.8))
])

gb_grid = GridSearchCV(gb_pipeline, gb_params, scoring="roc_auc", cv=3, n_jobs=-1)
gb_grid.fit(X_train_combo, y_train_combo)
print("Best GB params:", gb_grid.best_params_)

y_prob_gb = gb_grid.best_estimator_.predict_proba(X_test_combo)[:, 1]
y_pred_gb = gb_grid.best_estimator_.predict(X_test_combo)
print("[GradientBoosting combo] report:\n", classification_report(y_test_combo, y_pred_gb, digits=3))
print("[GradientBoosting combo] ROC-AUC:", roc_auc_score(y_test_combo, y_prob_gb))

```

### Step 6. Residual-based Target (log engagement - log followers)


```python
df["engagement_residual_alt"] = np.log1p(df["engagement_rate"].clip(lower=0) + 1e-9) - df["log_followers"]
res_cutoff = df["engagement_residual_alt"].quantile(0.75)
df["residual_high"] = (df["engagement_residual_alt"] >= res_cutoff).astype(int)
print({"residual_high 비율": df["residual_high"].mean(), "cutoff": res_cutoff})

X_res = df[feature_cols + ["combined_text"]]
y_res = df["residual_high"]

X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(
    X_res, y_res, test_size=0.25, stratify=y_res, random_state=42
)

res_pipeline = Pipeline([
    ("preprocess", combo_preprocessor),
    ("clf", HistGradientBoostingClassifier(class_weight="balanced", max_depth=8, random_state=42))
])
res_pipeline.fit(X_train_res, y_train_res)
y_prob_res = res_pipeline.predict_proba(X_test_res)[:, 1]
y_pred_res = res_pipeline.predict(X_test_res)

print("[Residual target HGB] report:\n", classification_report(y_test_res, y_pred_res, digits=3))
print("[Residual target HGB] ROC-AUC:", roc_auc_score(y_test_res, y_prob_res))

```

### Step 7. Sentence-BERT Embeddings + XGBoost


```python
train_idx = X_train_combo.index
test_idx = X_test_combo.index

text_model = SentenceTransformer("all-MiniLM-L6-v2")
train_emb = text_model.encode(df.loc[train_idx, "combined_text"].tolist(), batch_size=64, show_progress_bar=True)
test_emb = text_model.encode(df.loc[test_idx, "combined_text"].tolist(), batch_size=64, show_progress_bar=True)

num_scaler = StandardScaler().fit(df.loc[train_idx, feature_cols])
train_num = num_scaler.transform(df.loc[train_idx, feature_cols])
test_num = num_scaler.transform(df.loc[test_idx, feature_cols])

X_train_bert = np.hstack([train_num, train_emb])
X_test_bert = np.hstack([test_num, test_emb])

y_train_bert = y_train_combo
scale_pos_weight = (len(y_train_bert) - y_train_bert.sum()) / y_train_bert.sum()

xgb_clf = XGBClassifier(
    n_estimators=800,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="auc",
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    n_jobs=-1,
)
xgb_clf.fit(X_train_bert, y_train_bert)

y_prob_xgb = xgb_clf.predict_proba(X_test_bert)[:, 1]
y_pred_xgb = xgb_clf.predict(X_test_bert)
print("[XGBoost + SBERT] report:\n", classification_report(y_test_combo, y_pred_xgb, digits=3))
print("[XGBoost + SBERT] ROC-AUC:", roc_auc_score(y_test_combo, y_prob_xgb))

```

### Step 8. Simple Stacking (Logistic + GradientBoosting + XGBoost)


```python
X_caption = df[feature_cols + ["caption_text_clean"]]
X_train_caption = X_caption.loc[train_idx]
X_test_caption = X_caption.loc[test_idx]

log_stack_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("clf", LogisticRegression(max_iter=3000, class_weight="balanced", solver="saga"))
])
log_stack_pipeline.fit(X_train_caption, y_train_combo)
log_prob_train = log_stack_pipeline.predict_proba(X_train_caption)[:, 1]
log_prob_test = log_stack_pipeline.predict_proba(X_test_caption)[:, 1]

gb_prob_train = gb_grid.best_estimator_.predict_proba(X_train_combo)[:, 1]
# y_prob_gb already computed on test set as gb predictions
gb_prob_test = y_prob_gb

xgb_prob_train = xgb_clf.predict_proba(X_train_bert)[:, 1]
xgb_prob_test = y_prob_xgb

stack_train = np.vstack([log_prob_train, gb_prob_train, xgb_prob_train]).T
stack_test = np.vstack([log_prob_test, gb_prob_test, xgb_prob_test]).T

stack_meta = LogisticRegression(max_iter=2000, class_weight="balanced")
stack_meta.fit(stack_train, y_train_combo)
y_prob_stack = stack_meta.predict_proba(stack_test)[:, 1]
y_pred_stack = stack_meta.predict(stack_test)

print("[Stacked meta-model] report:\n", classification_report(y_test_combo, y_pred_stack, digits=3))
print("[Stacked meta-model] ROC-AUC:", roc_auc_score(y_test_combo, y_prob_stack))

```
